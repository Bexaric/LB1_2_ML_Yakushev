```python
df = pd.read_csv('D:/My files/Учёба и игры и т.д/Магистр/2_семестр_учёба/МетодыИИ/LB1_2_ML_Yakushev/notebook/EDA.ipynb')
df.head()
df.info()
display(df.describe())
```

```python
print(df['Country'].unique())
print(df['Status'].unique())
```

```python
from sklearn.preprocessing import LabelBinarizer, OneHotEncoder

df_old = df

lb = LabelBinarizer()
df['Status'] = lb.fit_transform(df['Status'])

for col in ['Country']:
    ohe = OneHotEncoder()
    encoded = ohe.fit_transform(df[[col]]).toarray()
    df = df.join(pd.DataFrame(encoded, columns=[f"{col}_{c}" for c in ohe.categories_[0]], index=df.index))
    df.drop(col, axis=1, inplace=True)

display(df, df_old)
```

```python
# Гистограммы
num_columns = df.select_dtypes(include=np.number).columns.tolist()
n_cols = len(num_columns)
fig, axes = plt.subplots(n_cols, 1, figsize=(15, n_cols*3), sharex=False)
for ax, col in zip(axes, num_columns):
    ax.hist(df[col].dropna(), bins=50, edgecolor='black')
    ax.set_title(col)
plt.tight_layout()
plt.show()
```

```python
# Тепловая карта корреляций
plt.figure(figsize=(20, 16))
sns.heatmap(df[num_columns].corr(), 
            annot=True, 
            cbar=True,
            cmap='coolwarm', 
            center=0,
            fmt='.2f',
            square=True,
            linewidths=0.5,
            linecolor='black')
plt.title('Матрица корреляций')
plt.show()
```
