```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import yaml
import json
import os
import joblib

from sklearn.preprocessing import LabelBinarizer, OneHotEncoder, MinMaxScaler
```

### <font color='#F5DEB3'>1.2 Чтение и анализ данных

```python
df = pd.read_csv('D:/My files/Учёба и игры и т.д/Магистр/2_семестр_учёба/МетодыИИ/LB1_2_ML_Yakushev/data/life_expectancy_data.csv')
display(df.head())
```

```python
display(df ,df.iloc[0])
```

```python
display(df.info())
```

```python
display(df.describe()) 
```

```python
display(df.duplicated().sum(), df.isnull().sum())
```

```python
print(df['Country'].unique())
print(df['Status'].unique())
```

```python
# Гистограммы
num_columns = df.select_dtypes(include=np.number).columns.tolist()
n_cols = len(num_columns)
fig, axes = plt.subplots(n_cols, 1, figsize=(15, n_cols*3), sharex=False)
for ax, col in zip(axes, num_columns):
    ax.hist(df[col].dropna(), bins=50, edgecolor='black')
    ax.set_title(col)
plt.tight_layout()
plt.show()
```

```python
# Тепловая карта корреляций
plt.figure(figsize=(20, 16))
sns.heatmap(abs(df[num_columns].corr()), 
            annot=True, 
            cbar=True,
            cmap='rocket_r', 
            center=0,
            fmt='.0%',
            annot_kws={'size':str(12)},
            square=True,
            linewidths=0.5,
            linecolor='black')
plt.title('Матрица корреляций', fontdict={'fontsize': 24, 'fontweight': 'bold'})
plt.show()
```

### <font color='#F5DEB3'>1.3 Обработка данных

```python
df.drop('Country', axis=1, inplace=True)
```

```python
lb = LabelBinarizer()
df['Status'] = lb.fit_transform(df['Status'])
df
```

```python
df.drop(['Year', 'Total expenditure', 'Population', 'Measles', 
         'Hepatitis B', 'under-five deaths', 'infant deaths'],axis=1, inplace=True)
df
```

```python
df.drop(['percentage expenditure', 'thinness 5-9 years', 
         'Income composition of resources'], axis=1, inplace=True)
```

```python
initial_rows = len(df)
df.dropna(inplace=True)
print(f"Удалено строк с пропусками: {initial_rows - len(df)}")
print(f"Осталось строк: {len(df)}")
```

```python
target_col = 'Life expectancy'
feature_cols = [col for col in df.columns if col != target_col]

status_col = 'Status'
other_features = [col for col in feature_cols if col != status_col]

scaler = MinMaxScaler(feature_range=(-1, 1))
df[other_features] = scaler.fit_transform(df[other_features])
```

### <font color='#F5DEB3'>1.4 Отображение данных

```python
display(df)
```

```python
# Гистограммы
num_columns = df.select_dtypes(include=np.number).columns.tolist()
n_cols = len(num_columns)
fig, axes = plt.subplots(n_cols, 1, figsize=(15, n_cols*3), sharex=False)
for ax, col in zip(axes, num_columns):
    ax.hist(df[col].dropna(), bins=50, edgecolor='black')
    ax.set_title(col)
plt.tight_layout()
plt.show()
```

```python
# Тепловая карта корреляций
plt.figure(figsize=(20, 16))
sns.heatmap(abs(df[num_columns].corr()), 
            annot=True, 
            cbar=True,
            cmap='rocket_r', 
            center=0,
            fmt='.0%',
            annot_kws={'size':str(12)},
            square=True,
            linewidths=0.5,
            linecolor='black')
plt.title('Матрица корреляций', fontdict={'fontsize': 24, 'fontweight': 'bold'})
plt.show()
```
